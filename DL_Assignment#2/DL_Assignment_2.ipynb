```
%tensorflow_version 2.x

import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import random
from progressbar import progressbar
from time import time
```

```
mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = np.expand_dims(train_images, axis=-1)/255.
train_labels = np.int64(train_labels)
test_images = np.expand_dims(test_images, axis=-1)/255.
test_labels = np.int64(test_labels)
```

```
plt.figure(figsize=(10,10))
random_inds = np.random.choice(60000,36)
for i in range(36):
    plt.subplot(6,6,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    image_ind = random_inds[i]
    plt.imshow(np.squeeze(train_images[image_ind]), cmap=plt.cm.binary)
    plt.xlabel(train_labels[image_ind])
```

```
def visualize(hist):
  # Get training and test loss histories
  training_loss = hist.history['loss']
  # Create count of the number of epochs
  epoch_count = range(1, len(training_loss) + 1)
  # Visualize loss history
  plt.figure()
  plt.plot(epoch_count, training_loss, 'r--')
  plt.legend(['Training Loss'])
  plt.xlabel('Epoch')
  plt.ylabel('Loss')
  plt.show()
```

```
def visualizewithAcc(hist):
  # Get training and test acc histories
  training_acc = hist.history['accuracy']
  val_acc = hist.history['val_accuracy']
  # Create count of the number of epochs
  epoch_count = range(1, len(training_acc) + 1)
  # Visualize acc history
  plt.figure()
  plt.plot(epoch_count, training_acc, 'r--')
  plt.plot(epoch_count, val_acc, 'b-')
  plt.legend(['Training acc', 'Val acc'])
  plt.xlabel('Epoch')
  plt.ylabel('accuracy')
  plt.show()
```

```
def build_fc_model(Optimizer='adam',HLactiviation='sigmoid',LOSS='sparse_categorical_crossentropy'):
  fc_model = tf.keras.Sequential([
      tf.keras.layers.Flatten(), # To flatten the image input
      tf.keras.layers.Dense(128, activation=HLactiviation),
      tf.keras.layers.Dense(10, activation='softmax')
  ])
  fc_model.compile(optimizer=Optimizer,
              loss=LOSS,
              metrics=['accuracy'])
  return fc_model
```

```
def build_cnn_model(Optimizer='adam',HLactiviation='sigmoid',LOSS='categorical_crossentropy',con_filter_size1=24,con_filter_size2=36,pooling_size=2,stride_length=1):
    cnn_model = tf.keras.Sequential([
      tf.keras.layers.Conv2D(con_filter_size1, (3,3),strides=(stride_length,stride_length) ,activation=HLactiviation, input_shape=(28,28,1)),
      tf.keras.layers.MaxPooling2D((pooling_size,pooling_size)),
      tf.keras.layers.Conv2D(con_filter_size2, (3,3), strides=(stride_length,stride_length),activation=HLactiviation),
      tf.keras.layers.MaxPooling2D((pooling_size,pooling_size)),
      tf.keras.layers.Flatten(), # To flatten the image input
      tf.keras.layers.Dense(784, activation=HLactiviation),
      tf.keras.layers.Dense(128, activation=HLactiviation),
      tf.keras.layers.Dense(10, activation='softmax')
    ])
    cnn_model.compile(optimizer=Optimizer,
              loss=LOSS,
              metrics=['accuracy'])
    return cnn_model
```

```
BATCH_SIZE = 64
EPOCHS = 5
optimizers = ['SGD', 'RMSprop', 'Adagrad', 'Adam', 'Adamax']
activiation = ['sigmoid','tanh','relu']
losses = ['categorical_crossentropy','mean_squared_error','mean_squared_logarithmic_error','binary_crossentropy','sparse_categorical_crossentropy']
#print(train_images.shape,train_labels.shape,test_images.shape,test_labels.shape)
for op in optimizers:
  for act in activiation:
    for loss in losses:
      model = build_fc_model(op,act,loss)
      history = model.fit(train_images, train_labels, epochs=EPOCHS,batch_size=BATCH_SIZE,verbose=0)
      test_loss, test_acc = model.evaluate(test_images,  test_labels,batch_size=BATCH_SIZE,verbose=0)
      visualize(history)
      print('Model (',op,act,loss,') with Test accuracy:', test_acc)
```

```
BATCH_SIZE = 64
EPOCHS = 5
optimizers = ['RMSprop','Adam']
activiation = ['tanh','relu']
losses = ['sparse_categorical_crossentropy']
for op in optimizers:
  for act in activiation:
    for loss in losses:
      start = time()
      model = build_cnn_model(op,act,loss,None,None,None,None)
      history = model.fit(train_images, train_labels, validation_data=(test_images, test_labels),epochs=EPOCHS,batch_size = BATCH_SIZE,verbose=0)
      test_loss, test_acc = model.evaluate(test_images,  test_labels,batch_size=BATCH_SIZE,verbose=0)
      end = time()
      visualizewithAcc(history)
      print('Model (',op,act,loss,') with Test accuracy:', test_acc,'and Test loss: ',test_loss,'In time: ',end-start)
```

```
start = time()
model = build_cnn_model(Optimizer='Adam',HLactiviation='relu',LOSS='sparse_categorical_crossentropy',stride_length=2)
history = model.fit(train_images, train_labels, validation_data=(test_images, test_labels),epochs=EPOCHS,batch_size = BATCH_SIZE,verbose=0)
test_loss, test_acc = model.evaluate(test_images,  test_labels,batch_size=BATCH_SIZE,verbose=0)
end = time()
visualizewithAcc(history)
print('Model (filter_size:24,36',',strides:',2,',pooling_size:',2,') with Test accuracy:', test_acc,'and Test loss: ',test_loss,'In time: ',end-start)

start = time()
model = build_cnn_model(Optimizer='Adam',HLactiviation='relu',LOSS='sparse_categorical_crossentropy',pooling_size=1)
history = model.fit(train_images, train_labels, validation_data=(test_images, test_labels),epochs=EPOCHS,batch_size = 2*BATCH_SIZE,verbose=2)
test_loss, test_acc = model.evaluate(test_images,  test_labels,batch_size=BATCH_SIZE,verbose=0)
end = time()
visualizewithAcc(history)
print('Model (filter_size:24,36',',strides:',1,',pooling_size:',1,') with Test accuracy:', test_acc,'and Test loss: ',test_loss,'In time: ',end-start)

predictions = model.predict(test_images)
import numpy
print(predictions[0] )
print(numpy.where(predictions[0] == max(predictions[0])),test_labels[0])
plt.imshow(np.squeeze(test_images[0]), cmap=plt.cm.binary)
```
