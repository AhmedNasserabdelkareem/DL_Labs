```
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Embedding
from keras.layers import LSTM,GRU,SimpleRNN
from keras.datasets import imdb
```

```
#BASIC RNN
modelbase = Sequential()
modelbase.add(Embedding(100000, 64))
modelbase.add(SimpleRNN(128, dropout=0.2, recurrent_dropout=0.2))
modelbase.add(Dense(1, activation='sigmoid'))
#GRU
modelgru = Sequential()
modelgru.add(Embedding(100000, 64))
modelgru.add(GRU(128, dropout=0.2, recurrent_dropout=0.2))
modelgru.add(Dense(1, activation='sigmoid'))
#LSTM
modellstm = Sequential()
modellstm.add(Embedding(100000, 64))
modellstm.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
modellstm.add(Dense(1, activation='sigmoid'))

models = [modelbase,modelgru,modellstm]
labels = ["Basic RNN Model","GRU Model","LSTM Model"]
```

```
def visualize(hist):
  # Get training and test acc histories
  training_acc = hist.history['accuracy']
  val_acc = hist.history['val_accuracy']
  # Create count of the number of epochs
  epoch_count = range(1, len(training_acc) + 1)
  # Visualize acc history
  plt.figure()
  plt.plot(epoch_count, training_acc, 'r--')
  plt.plot(epoch_count, val_acc, 'b-')
  plt.legend(['Training acc', 'Val acc'])
  plt.xlabel('Epoch')
  plt.ylabel('accuracy')
  plt.show()
```

```
def build_model(model,x_train,x_test,y_train,y_test):
  model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
  hist = model.fit(x_train, y_train,batch_size=64,epochs=10,validation_data=(x_test, y_test))
  score, acc = model.evaluate(x_test, y_test,batch_size=32)
  return hist,score,acc
```

```
from keras.datasets import imdb
(x_train, y_train), (x_test, y_test) = imdb.load_data(path="imdb.npz",num_words=None,skip_top=0,maxlen=None,seed=113,start_char=1,oov_char=2,index_from=3)
x_train = sequence.pad_sequences(x_train, maxlen=80)
x_test = sequence.pad_sequences(x_test, maxlen=80)
count = 0
for model in models:
  hist,sc,acc = build_model(model,x_train,x_test,y_train,y_test)
  print(labels[count])
  print("Accuracy: ",acc)
  visualize(hist)
  count+=1
```
